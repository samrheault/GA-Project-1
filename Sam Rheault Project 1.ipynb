## Challenge 1: Largest Palindrome
A palindromic number reads the same both ways. For example, 1234321 is a palindrome. The largest palindrome made from the product of two two-digit numbers is 9009 = 91 × 99. Find the largest palindrome made from the product of two three-digit numbers. Afterward, write a brief explanation walking through your code's logic in markdown.

```python
def checkpalindrome(n):
#Checks if n is a palindrome
    palindrome = False
#first convert the input to a string
    n = str(n)
    count = 0
#check if the first character is the same as the last character, the second the same as second to last, etc...
    for index, element in enumerate(n):
        opposite = (len(n)-index)-1
        if element == n[opposite]:
#if the character we are looking at is the same as its opposite, increase the counter
            count += 1
#if all of the characters are the same as their opposites, the counter should be equal to the string length
    if count == len(n):
        palindrome = True
    return palindrome

palindromes = {}
#check the product of every 3 digit number (100-999) with every other 3 digit number
for one in range(100, 999):
    for two in range(100, 999):
        product = one*two
#if the product is a palindrome, store the product and the two numbers that created it in a dictionary
        if checkpalindrome(product):
            parents = (one, two)
            palindromes[product]=parents
#the largest product is determined and printed along with the two numbers that were multiplied to create it
largest = max(palindromes)
print("The largest palindromic number is", largest)
print(largest, "is a product of", palindromes[largest])
```

## Challenge 2: Summation of Primes
The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17. Find the sum of all the primes below 2,000. Afterward, write a brief explanation walking through your code's logic in markdown.

```python
sumofprimes = 0
#checking every number under 2000
for i in range (2000):
    isprime = True
    for j in range (2, i):
#dividing each number under 2000 by each other number lower than itself (besides 1)
        if i%j == 0:
#if any number is evenly divisible by any number other than 1 or itself, it cannot be prime
            isprime = False
    if isprime == True:
#if the number is not evenly divisible by any other number, add it to the sum variable
        sumofprimes += i
print("The sum of all prime numbers below 2000 is", sumofprimes)
```

## Challenge 4: String Compressor
Implement a method to perform basic string compression using the counts of repeated characters. (This is called run-length encoding.) For example, the string "aabcccccaaa" would become a2b1c5a3. If the “compressed” string would not become smaller than the original string, your method should return the original string. You can assume the string has only uppercase and lowercase letters (a–z). Specify whether your solution is case sensitive or case insensitive and what you would need to change to make it the other. Afterward, write a brief explanation walking through your code's logic in markdown.

```python
def compress(string):
    counter = 1
    newstring = ""
    for i in range(len(string)-1):
#checks if each element in the string is the same as the next
        if string[i] == string[i+1]:
#if it is the same, count the number of occurances of that element            
            counter +=1
        else:
#if it is not the same, add the element and its count to the compressed string and reset the counter to move on
            newstring = newstring + string[i]+ str(counter)
            counter = 1
#add the last elements to the compressed string (because the last element cannot be the same or different as the next, there is no next)
    newstring = newstring + string[i]+ str(counter)
    return (newstring)

#ask the user to type a string to compress. Lowercase and uppercase versions of the same letter will be considered different characters
#to make it case insensitive:
#test = input("type a string to be compressed. Compression is not case sensitive\n").lower()
test = input("type a string to be compressed. Compression is case sensitive\n")
#if the compressed string is shorter than the original, print the original and the compressed version
if len(compress(test))< len(test):
    print("original string: ", test)
    print("compressed string: ", compress(test))
#if the compressed string is not shorter, print a message saying so
else:
    print("compressing the string would not shorten it")
```

## BONUS Challenge: Martingal Betting Strategy for Roulette
An 'interesting' strategy to beat the house at Roulette is the Martingal Betting Strategy. You can see a brief description here: https://www.roulettesites.org/strategies/martingale/.

Placing Bets & Doubling Up You start with a small amount, preferably the table minimum, and keep betting the same until you lose. When this occurs, double the size of your bet for the next spin. This way, in case you win, you will recover the money you lost on the previous round, and win something extra. If you keep on losing, keep on doubling your bet – the logic stays the same. As soon as you win, you should restart and bet the smallest amount for the next spin. Rinse and repeat.

Write a program that simulates a person applying this strategy who starts with an initial amount of 10k. To keep things simple, assume that one of the simple bets is used each time (e. g. red, black, odd or even) as these bets have probability of success of 16/38 and probability of failure 18/38. The player starts with an  1bet.Ifthebetissuccesful,the 1 bet is used again. If the player loses the  1bet,a 2 bet is used next time. If still unsuccessful, the bet doubles each time for  4, 8, etc. Upon winning the bet, next bet is reset to $1. Use the code to get intuition on how this strategy would fare in real life. Afterward, write a brief explanation walking through your code's logic in markdown.

```python
import random
#starting conditions: bet is 1 and you have $10K, 0 wins and 0 losses
bet = 1
balance = 10000
wins = 0
losses = 0
#"Tries" represents the number of spins you want to try this strategy for
tries = 100

for i in range(tries):
#create a random number between 1 and 36, the odds of it being 1-16 are the same as the odds of winning on one of the simple bets    
    if random.randint(1, 36) <=16:
#your balance increases by the amount you bet
        balance += bet
        print("you won ", bet)
#add 1 to the wins counter        
        wins +=1
#reset the bet to 1        
        bet = 1
        
    else:
#the odds of the number being 17-36 are the same as the odds of losing
        print("you lost ", bet)
#deduct the amount you bet from your balance        
        balance = balance - bet
#multiply the bet by 2 for next time
        bet = bet*2
#add 1 to the losses counter        
        losses +=1
#prints your balance after each spin        
    print("balance: ", balance)
#print the summary
print ("You tried", tries, "times. You won", wins, "times and lost", losses, "times. Your balance increased by", (balance-10000))
print ("Your final balance is", balance)
#This strategy ultimately makes you money, but takes a lot of spins to get significant earnings
#The number of losses is almost always greater than the number of wins, but the earnings are always positive
```
